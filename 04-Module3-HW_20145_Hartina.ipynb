# Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 1:
- Read the boston house dataset

- Create train and test datasets. 

```python
import pandas as pd
df = pd.read_csv('boston_house_prices.csv')
df.head()
df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('MEDV', axis=1)
y = df['MEDV']
X.head()
```

```python
X.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape,y_train.shape, y_test.shape
```

### Step 2:
Create a scatter plot of each attribute with the mean house price.

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(X_train['RM'], y_train)
```

```python
for i, col in enumerate(X_train.columns):
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    ax.scatter(X_train[col], y_train)
    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel('MEDV')
```

```python
plt.tight_layout()
plt.show()
```

### Step 3:
Create a box plot of all attributes.

```python
plt.Figure(figsize=(10, 10))
df.boxplot(rot=45)
plt.title('Boxplot of all features')
plt.ylabel('Values Range')
plt.show()
```

### Step 4:
Use `StandardScaler()` to scale the trainin data set.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled[:5]
```

### Step 5:
Use `KNeighborsRegressor()` to fit both not-scaled and scaled datasets. Check the scores of test dataset for both models.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
```

```python

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn.fit(X_train, y_train)
y_pred_original = knn.predict(X_test) 
score_original = r2_score(y_test, y_pred_original)
```

```python
knn.fit(X_train_scaled, y_train)
y_pred_scaled = knn.predict(X_test_scaled)
score_scaled = r2_score(y_test, y_pred_scaled)
```

```python
print(f"R² score (Not Scaled): {score_original:.4f}")
print(f"R² score (Scaled): {score_scaled:.4f}")
```

### Step 6:
Repeat Step 5 for `RandomForestRegressor`. Use

`RandomForestRegressor(n_estimators=100, random_state=0)`

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
rf.fit(X_train, y_train)
y_pred_original = rf.predict(X_test)
score_original = r2_score(y_test, y_pred_original)
```

```python
rf.fit(X_train_scaled, y_train)
y_pred_scaled = rf.predict(X_test_scaled)
score_scaled = r2_score(y_test, y_pred_scaled)
```

```python
print(f"R² score (Not Scaled): {score_original:.4f}")
print(f"R² score (Scaled): {score_scaled:.4f}")
```
