```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
heart = pd.read_csv('Heart.csv')

# Force the response into a binary indicator:
heart['AHD'] = 1*(heart['AHD'] == "Yes")

print(heart.shape)
#heart.head()
heart.describe()
```

```python
heart_train, heart_test = train_test_split(heart, test_size=0.3, random_state = 109)
```

**Q1.1** Below we fit an unregularized logistic regression model (`logit1`) to predict `AHD` from `Age` and `MaxHR` in the training set (with `penalty='none'`).  Print out the coefficient estimates, and interpret general trends.

```python
degree = 1
predictors = ['Age','MaxHR']

X_train1 = PolynomialFeatures(degree=degree, include_bias=False).fit_transform(heart_train[predictors])
y_train = heart_train['AHD']

logit1 = LogisticRegression(max_iter=5000, penalty=None).fit(X_train1, y_train)

print("Logistic Regression Estimated Betas:",
      logit1.coef_, logit1.intercept_)
```

*your interpretation here*

```python
The logistic regression model `logit1` was fitted to predict the presence of AHD (Atherosclerotic Heart Disease) based on two predictors: Age and MaxHR (Maximum Heart Rate). The estimated coefficients (betas) from the model indicate the relationship between each predictor and the likelihood of having AHD.

- The coefficient for Age is positive, suggesting that as age increases, the likelihood of having AHD also increases.
- The coefficient for MaxHR is negative, indicating that higher maximum heart rates are associated with a lower likelihood of having AHD.

These trends align with general medical knowledge, where older age is a risk factor for heart disease, and higher physical fitness (often indicated by higher MaxHR) is associated with a lower risk of heart disease.
```

**Q1.3** Compare the two models above on how they create the classification boundary.  Which is more likely to be overfit?  How would regularization affect these boundaries?

The model with the 1st order polynomial (degree=1) will create a linear classification boundary, while the model with the 4th order polynomial (degree=4) will create a more complex, non-linear boundary.

**Overfitting:**

The 4th order polynomial model is more likely to be overfit because it has more flexibility to fit the training data closely, capturing noise and fluctuations that do not generalize well to new data.
The 1st order polynomial model is less likely to overfit because it has a simpler, linear boundary that may not capture all the nuances of the training data but is more likely to generalize well.
Effect of Regularization:

**Regularization** (such as L1 or L2 regularization) would penalize the complexity of the model, effectively reducing the coefficients of less important features.
For the 4th order polynomial model, regularization would help to prevent overfitting by discouraging the model from fitting the noise in the training data.
For the 1st order polynomial model, regularization would have a smaller effect since the model is already simple, but it could still help to improve generalization by ensuring that the model does not rely too heavily on any single feature.
